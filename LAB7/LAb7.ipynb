```python
from PIL import Image, ImageDraw
input_image = Image.open('Screenshot 2025-07-03 133935.png')
input_pixels = input_image.load()
box_kernel = [[1 / 9, 1 / 9, 1 / 9],
              [1 / 9, 1 / 9, 1 / 9],
              [1 / 9, 1 / 9, 1 / 9]]

kernel = box_kernel
offset = len(kernel) // 2

output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(kernel)):
            for b in range(len(kernel)):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                acc[0] += pixel[0] * kernel[a][b]
                acc[1] += pixel[1] * kernel[a][b]
                acc[2] += pixel[2] * kernel[a][b]
        draw.point((x, y), (int(acc[0]), int(acc[1]), int(acc[2])))

output_image.save('traidat_lammo.PNG')
```

```python
import numpy as np
from PIL import Image, ImageDraw

# Phần code cho bài 1: Làm mờ ảnh bằng Gaussian Filter
input_image = Image.open('Screenshot 2025-07-03 133935.png')
input_pixels = input_image.load()

gaussian_kernel = [
    [1 / 256, 4 / 256, 6 / 256, 4 / 256, 1 / 256],
    [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
    [6 / 256, 24 / 256, 36 / 256, 24 / 256, 6 / 256],
    [4 / 256, 16 / 256, 24 / 256, 16 / 256, 4 / 256],
    [1 / 256, 4 / 256, 6 / 256, 4 / 256, 1 / 256]
]

kernel = gaussian_kernel
offset = len(kernel) // 2

output_image = Image.new("RGB", input_image.size)
draw = ImageDraw.Draw(output_image)

for x in range(offset, input_image.width - offset):
    for y in range(offset, input_image.height - offset):
        acc = [0, 0, 0]
        for a in range(len(kernel)):
            for b in range(len(kernel[0])):
                xn = x + a - offset
                yn = y + b - offset
                pixel = input_pixels[xn, yn]
                acc[0] += pixel[0] * kernel[a][b]
                acc[1] += pixel[1] * kernel[a][b]
                acc[2] += pixel[2] * kernel[a][b]
        
        r = int(max(0, min(255, acc[0])))
        g = int(max(0, min(255, acc[1])))
        b = int(max(0, min(255, acc[2])))
        
        draw.point((x, y), (r, g, b))

output_image.save('gaussian_blurred_image.PNG')

print("Ảnh đã được làm mờ bằng Gaussian Filter và lưu tại 'gaussian_blurred_image.PNG'")

# Phần code cho bài 2: Tính toán PageRank
# Giả định ma trận A là 4x4 để khớp với việc có x_D trong bảng và phép nhân Ax hợp lệ.
# Cột thứ 4 được thêm vào với giả định để đảm bảo kích thước phù hợp.
A = np.array([[0.0, 0.5, 0.5, 0.0],
              [1.0/3, 0.0, 0.5, 0.0],
              [1.0/3, 1.0, 0.0, 0.0],
              [1.0/3, 0.0, 0.0, 1.0]])

x = np.array([[1.0],
              [1.0],
              [1.0],
              [1.0]])

print("\n--- Tính toán PageRank ---")
print("Lần lặp 0:")
print(f"xA: {x[0,0]:.6f}")
print(f"xB: {x[1,0]:.6f}")
print(f"xC: {x[2,0]:.6f}")
print(f"xD: {x[3,0]:.6f}")
print("-" * 30)

for i in range(1, 11):
    x_new = np.dot(A, x)
    x = x_new
    print(f"Lần lặp {i}:")
    print(f"xA: {x[0,0]:.6f}")
    print(f"xB: {x[1,0]:.6f}")
    print(f"xC: {x[2,0]:.6f}")
    print(f"xD: {x[3,0]:.6f}")
    print("-" * 30)
```
